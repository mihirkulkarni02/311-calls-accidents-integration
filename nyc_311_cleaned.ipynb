```python
import pandas as pd
```

```python
df = pd.read_csv('nyc_311.csv')

#renames zip codes to be one word
df = df.rename(columns={"Zip Codes" : "Zipcodes"})

df['Created Date'] = pd.to_datetime(df['Created Date'])

df['Date'] = df['Created Date'].dt.date

df['Time'] = df['Created Date'].dt.time

df = df.drop('Created Date', axis =1)

df
```

```python
#specifies zip codes to manhattan only
df = df[df.notnull().all(axis = 1)]
df = df[df['Zipcodes'] > 10000]
df = df[df['Zipcodes'] < 10283]
df
```

```python
df.to_csv("final_311.csv")
```

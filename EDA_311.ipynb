```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the dataset
file_path = 'final_311.csv'
data = pd.read_csv(file_path)

# Displaying the first few rows of the dataset to understand its structure
data.head()
```

```python
# Visualizing the distribution of complaint types
plt.figure(figsize=(10, 6))
data['Complaint Type'].value_counts().head(10).plot(kind='barh')
plt.title('Top 10 Complaint Types')
plt.xlabel('Frequency')
plt.ylabel('Complaint Type')
plt.tight_layout()
plt.show()

# Converting 'Date' column to datetime for time series analysis
data['Date'] = pd.to_datetime(data['Date'])

# Frequency of complaints over time
plt.figure(figsize=(12, 6))
data['Date'].value_counts().sort_index().plot()
plt.title('Frequency of Complaints Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Complaints')
plt.tight_layout()
plt.show()

```
